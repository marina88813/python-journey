```python
import pandas as pd
import matplotlib.pyplot as plt

# Данные о том, сколько папок мы только что удалили
data = {
    'Компонент Xbox': ['Удалено (Game Bar, TCUI, Apps)', 'Оставлено (IdentityProvider)'],
    'Количество папок': [14, 3]
}

# Загружаем данные в таблицу pandas
df = pd.DataFrame(data)

# Строим красивую круговую диаграмму с помощью matplotlib
plt.figure(figsize=(6, 6))
plt.pie(
    df['Количество папок'], 
    labels=df['Компонент Xbox'], 
    autopct='%1.1f%%', 
    colors=['#ff4d4d', '#2ecc71'], 
    startangle=140
)
plt.title('Результат очистки папки WindowsApps от компонентов Xbox')

# Выводим график на экран
plt.show()
```
