```python
import math
from ColabTurtlePlus import Turtle as turtle

turtle.clearscreen()
turtle.setup(500, 500)
turtle.bgcolor("black")

t = turtle.Turtle()
t.hideturtle()
t.speed(13)
t.width(2)

def heart(scale, color):
    t.color(color)
    t.begin_fill()
    for i in range(360):
        x = scale * 16 * math.sin(math.radians(i)) ** 3
        y = scale * (
            13 * math.cos(math.radians(i))
            - 5 * math.cos(2 * math.radians(i))
            - 2 * math.cos(3 * math.radians(i))
            - math.cos(4 * math.radians(i))
        )
        t.goto(x, y)
    t.end_fill()

for s in range(12, 8, -1):
    t.penup()
    t.goto(0, 0)
    t.pendown()
    heart(s, "#b2dfdb")

t.penup()
t.goto(0, 0)
t.pendown()
heart(8, "#80cbc4")
```
